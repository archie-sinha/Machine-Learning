```python
#Create a new data frame, by replacing “??” with NAN
data_frame_1 = data_frame.replace(['??','????'], 'NAN')
data_frame_1.head(10)
#OR
cars_data=pd.read_csv("Toyota.csv",na_values=["??", "????"])
cars_data.head(10)
```

```python
#Replace the categorical values in the “Doors” column with its corresponding numeric value

cars_data["Doors"].replace("three",3, inplace=True)
cars_data["Doors"].replace("four",4, inplace=True)
cars_data["Doors"].replace("five",5, inplace=True)
cars_data.head(10)
```

```python
cars_data.info()
```

```python
#Convert data types of columns “Doors”, “MetColor” and “Automatic” to int, and object
cars_data["Doors"]= cars_data["Doors"].astype('int64')
cars_data["MetColor"]= cars_data["MetColor"].astype('object')
cars_data["Automatic"]= cars_data["Automatic"].astype('object')
```

```python
#Drop rows with null values
cars_data= cars_data.dropna() #droping rows with NUll values
```

```python
#Identify the total number of null values in each column of the data set
cars_data.isnull().sum()
```

```python
#Identify total number of cars that runs on “Petrol”, “Diesel” or “CNG”
cars_data["FuelType"].value_counts() #works only if we have CATEGORICAL DATA
```

```python
#Identify mean of “KM” for the cars that runs on “Diesel”
cars_data[cars_data.FuelType=="Diesel"].KM.mean()
```
